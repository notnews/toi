```python
import pandas as pd
import janitor
from utilities import (
    check_make_dir,
    save_mpl_fig,
    plotline, 
)
check_make_dir('../data', '../figures')
FP = '../data/toi_yearly_minority_dist.csv'

df = (pd.read_csv(FP, index_col=0)
      .assign(
          muslim_dist=lambda df: 100*df.muslim_dist,
          women_dist=lambda df: 100*df.women_dist,
          dist_st=lambda df: 100*df.dist_st,
          dist_sc=lambda df: 100*df.dist_sc,
      )
     )
df.head(3)
```

```python
ax = plotline(x=df.index, y=df.muslim_dist, 
              bw=.2,
              xrange=range(1840, 2011, 10),
              yrange=range(2, 10, 2), 
              title='Muslim representation (%)', 
             )    
save_mpl_fig('../figures/minority_yearly_muslim')
```

```python
ax = plotline(x=df.index, y=df.women_dist, 
              bw=.2,
              xrange=range(1840, 2011, 10),
#               yrange=range(2, 10, 2), 
              title='Women representation (%)', 
             )    
save_mpl_fig('../figures/minority_yearly_women')
```

```python
ax = plotline(x=df.index, y=df.dist_stsc, 
              bw=.2,
              xrange=range(1840, 2011, 10),
              yrange=[0,.25,.5,.75, 1], 
              title='Lower caste representation (%)', 
             )    
ax.set_ylim(0,1)
save_mpl_fig('../figures/minority_yearly_caste')
```
